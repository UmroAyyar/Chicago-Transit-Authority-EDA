```python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
```

```python
path = 'bus_tracking-apr-300.csv'
bus_data = pd.read_csv(path)
```

```python
bus_data
```

```python
columns = ["vid", "tmstmp", "lat", "lon", "hdg", "pid", "rt","des","pdist","dly", "tablockid", "tatripid", "zone", "None"]

```

```python
bus_data.columns = columns
```

```python
bus_data
```

```python
bus_data = bus_data.drop(columns=["None"])
```

```python
bus_data
```

```python
bus_data[['date', 'time']] = bus_data['tmstmp'].str.split(' ', expand=True)

# Extracting year and month
bus_data['year'] = bus_data['date'].str[:4]
bus_data['month'] = bus_data['date'].str[4:]

# Dropping the intermediate 'date' column
bus_data.drop(columns=['date'], inplace=True)

# Renaming the 'time' column to 'timestamp' for clarity
bus_data.rename(columns={'time': 'timestamp'}, inplace=True)

# Printing the modified DataFrame
print(bus_data)
```

```python
des_value_counts = bus_data["rt"].value_counts()

print(des_value_counts)
```

```python
subset_data = bus_data[(bus_data["rt"] == '3')]
print(subset_data)
```

```python
subset_data
```

```python
pid_18414 = subset_data[(bus_data["pid"] == 18414)]
```

```python
pid_18414
```

```python
april_10 = pid_18414[(pid_18414["month"] == '0410')]
april_10
```

```python
hdg_value_counts = april_10["hdg"].value_counts()

print(hdg_value_counts)
```

```python
hdg_1 = april_10[(april_10["hdg"] == 1)]
hdg_1
```

```python
hdg_1['timestamp'] = pd.to_timedelta(hdg_1['timestamp'])

# Calculate the difference between each row and the row above
time_diff = hdg_1['timestamp'].diff()

# Convert time difference to seconds for comparison
time_diff_seconds = time_diff.dt.total_seconds()

# Mark rows where the time difference is less than or equal to 2 minutes as 1 in a new column 'bunched'
hdg_1['bunched'] = (time_diff_seconds <= 120).astype(int)

# Print the modified DataFrame
print(hdg_1)
```

```python
hdg_1
```

```python
hdg_1_value_counts = hdg_1["bunched"].value_counts()

print(hdg_1_value_counts)
```

```python
subset_data = bus_data.copy()  # Make a copy of the original DataFrame

# Step 2: Subset for all values of "pid"
pid_data = subset_data.copy()  # Make a copy of the subset data
pid_data = pid_data.astype({"pid": str})  # Convert "pid" column to string for comparison
#print(pid_data)

# Step 3: Subset for all values of "month"
month_data = pid_data.copy()  # Make a copy of the subset data
month_data["month"] = month_data["month"]  # Extract month from timestamp
print(month_data)

# Step 4: Subset for all values of "hdg"
hdg_data = month_data.copy()  # Make a copy of the subset data
hdg_data = hdg_data.astype({"hdg": int})  # Convert "hdg" column to int for comparison
print(hdg_data)

# Step 5: Convert "timestamp" to timedelta
hdg_data['timestamp'] = pd.to_timedelta(hdg_data['timestamp'])

# Calculate the difference between each row and the row above
time_diff = hdg_data.groupby(["rt", "pid", "month", "hdg"])['timestamp'].diff()

# Convert time difference to seconds for comparison
time_diff_seconds = time_diff.dt.total_seconds()

# Mark rows where the time difference is less than or equal to 2 minutes as 1 in a new column 'bunched'
hdg_data['bunched'] = (time_diff_seconds <= 120).astype(int)

# Print the modified DataFrame
print(hdg_data)
```

```python
hdg_data
```

```python
hdg_data.to_csv( "hdg_data.csv")
```

```python
hdg_1_value_counts = hdg_data["month"].value_counts()

print(hdg_1_value_counts)
```

```python
hdg_1_value_counts = hdg_data["bunched"].value_counts()

# Plotting the bar chart
hdg_1_value_counts.plot(kind='bar')

# Adding labels and title
plt.xlabel('Bunched')
plt.ylabel('Count')
plt.title('Bunched Buses')

# Showing the plot
plt.show()
```

```python
month_bunched_counts = hdg_data.groupby("month")["bunched"].value_counts()

# Reshaping the data to make "month" values as columns
month_bunched_counts = month_bunched_counts.unstack().fillna(0)

# Plotting the bar chart
month_bunched_counts.plot(kind='bar', stacked=True)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Bunched Buses for Each Month')

# Showing the plot
plt.show()
```
